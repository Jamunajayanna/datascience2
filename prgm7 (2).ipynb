```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("/home/rajatraj/Downloads/Placement_Data_Full_Class.csv")
df
```

```python
df=df.drop(["Sno"],axis=1)
df=df.drop(["salary"],axis=1)
```

```python
x=df[['12thP','10thP','DegreeP','MbaP']]
y=df.status.values
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
y=le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB

nb=GaussianNB()
nb.fit(x_train,y_train)
```

```python
print('Naive Bayes Score' , nb.score(x_test,y_test)*100)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score,classification_report

prediction = nb.predict(x_test)

cm=confusion_matrix(y_test,prediction)
print(cm)
```

```python
accuracy = accuracy_score(y_test, prediction)
print(f'Accuracy: {accuracy*100}')
```

```python
precision = precision_score(y_test,prediction)
print(precision)
```

```python
# recall = recall(y_test,prediction)
```

```python
print(classification_report(y_test, prediction))
```

```python
from sklearn.naive_bayes import MultinomialNB

mNB=MultinomialNB()
mNB.fit(x_train,y_train)

print('MultinomialNB',mNB.score(x_test,y_test)*100)
```

```python
from sklearn.naive_bayes import BernoulliNB

bNB=BernoulliNB(binarize=0.0)
bNB.fit(x_train,y_train)
print('BernoulliNB',bNB.score(x_test,y_test)*100)
```
