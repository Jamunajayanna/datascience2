```python
import re
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer

from wordcloud import WordCloud
import matplotlib.pyplot as plt 
from nltk.probability import FreqDist

data = open("/home/rajatraj/Downloads/dataset/book.txt" ,'r').read()

sentences = sent_tokenize(data)
words= word_tokenize(data)
len(data)
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')
```

```python
word=[word.lower() for word in words if word.isalpha()]
```

```python
stop_words = set(stopwords.words('english'))
filter_words = [word for word in words if word not in stop_words]

stemmer = PorterStemmer()
stemmer_words=[stemmer.stem(word) for word in filter_words]
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words=[lemmatizer.lemmatize(word) for word in filter_words]
```

```python
processed_text = ''.join(lemmatized_words)
len(processed_text)
```

```python
words= nltk.word_tokenize(processed_text)
word_freq = nltk.FreqDist(words)

top_words= word_freq.most_common(40)
top_words
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
from PIL import Image
import numpy as np
```

```python
wordcloud= WordCloud(width=3000, height=2000,random_state=1, background_color="black",colormap='Pastel1',collocations=False, stopwords=STOPWORDS).generate_from_frequencies(dict(top_words))
```

```python
plt.figure(figsize=(40,30))
plt.imshow(wordcloud)
plt.axis("off")
```
